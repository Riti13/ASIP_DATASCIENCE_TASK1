```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/Video_Games_Sales_as_at_22_Dec_2016.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data=data.dropna(axis=0)
```

```python
data.isnull().sum()
```

```python
data=data.drop(columns=['Platform','Critic_Score','Critic_Count','User_Count','User_Score','Developer','Rating'],axis=1)
```

```python
data
```

```python
data=data[(data.Global_Sales >=10.0 ) & (data.Global_Sales <= 90.0)]
```

```python
data['Global_Sales'].value_counts()
```

```python
sns.set(style='darkgrid')
sns.countplot(y=data['Genre'])
```

```python
sns.regplot(y=data['Genre'],x=data['Global_Sales'])
```

```python
sns.distplot(x=data['Global_Sales'])
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rc=RandomForestRegressor()
rc.fit(xtrain,ytrain)
```

```python
pred=rc.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(pred,ytest))
rmse
```
